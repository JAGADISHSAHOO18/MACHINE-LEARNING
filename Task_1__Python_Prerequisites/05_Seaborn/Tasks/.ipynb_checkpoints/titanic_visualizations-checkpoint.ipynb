```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\jagad\Desktop\MACHINE_LEARNING\Titanic-Dataset.csv")
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Cabin'] = df['Cabin'].fillna("Unknown")
df = df.drop_duplicates()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x="Age", hue="Survived", bins=30)
plt.title("Survival Distribution by Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(data=df, x="Pclass", y="Survived", hue="Pclass", palette="Blues", legend=False)
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(data=df, x="Sex", y="Survived", hue="Sex", palette=["pink", "lightblue"], legend=False)
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Age", y="Survived", hue="Pclass", style="Sex", palette="coolwarm", alpha=0.6)
plt.title("Survival vs Age by Class and Gender")
plt.xlabel("Age")
plt.ylabel("Survived")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=df, x="Pclass", y="Age", hue="Survived")
plt.title("Age Distribution by Class and Survival")
plt.show()
```

```python
plt.figure(figsize=(8,6))
correlation = df[['Survived','Pclass','Age','Fare']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
