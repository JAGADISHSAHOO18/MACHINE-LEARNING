```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = pd.read_csv(r"C:\Users\jagad\Desktop\MACHINE_LEARNING\Titanic-Dataset.csv")
```

```python
print(a.head())
```

```python
print(a.isnull().sum())
```

```python
a = a.drop_duplicates()
a['Age'] = a['Age'].fillna(a['Age'].median())
a['Embarked'] = a['Embarked'].fillna(a['Embarked'].mode()[0])
a['Cabin'] = a['Cabin'].fillna("Unknown")
```

```python
class_survival = a.groupby('Pclass')['Survived'].mean()
gender_survival = a.groupby('Sex')['Survived'].mean()
age_survival = a.groupby('Age')['Survived'].mean()
```

```python
print(class_survival)
```

```python
print(gender_survival)
```

```python
print(age_survival)
```

```python
plt.figure(figsize=(15, 10))
```

```python
plt.subplot(2, 3, 1)
gender_survival.plot(kind='bar', color=['pink','blue'])
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.xlabel("Gender")
```

```python
plt.subplot(2, 3, 2)
age_survival.plot(kind='line', color='green')
plt.title("Survival Rate by Age Groups")
plt.ylabel("Survival Rate")
plt.xlabel("Age Groups")
```

```python
plt.subplot(2, 3, 3)
plt.scatter(a['Age'], a['Fare'], c=a['Survived'], cmap='coolwarm', alpha=0.6)
plt.title("Age vs Fare (Colored by Survival)")
plt.xlabel("Age")
plt.ylabel("Fare")
```

```python
plt.subplot(2, 3, 4)
plt.hist(a['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
```

```python
plt.subplot(2, 3, 5)
survival_count = a['Survived'].value_counts()
plt.pie(survival_count, labels=['Died', 'Survived'], autopct='%1.1f%%', colors=['red', 'green'])
plt.title("Survival Distribution")
```

```python
plt.tight_layout()
```

```python
plt.show()
```
